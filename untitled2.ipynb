```python
importimport numpy as np
import matplotlib.pyplot as plt
```

se hace uso de estas librerias porque librerias dado que la libreria numpy, es la encargada de realizar los calculos, mientras que la libreria matpit es la resposable de la grafica

```python
p = 0.5
n_pasos = 1000
q = 1 - p
```

en esta parte del programa le asignamos el valor que desiemos a p, al igual el numero de pasos

```python
pasos = np.random.choice([1, -1], size=n_pasos, p=[p, q])


trayectoria = np.cumsum(pasos)
trayectoria = np.insert(trayectoria, 0, 0)
```

esta parte del codigo es la encarga de darle valovalores a cada paso, es decir decide si cada en cada n-paso se avanza o se atraza, sisiendo representado estos valores como 1 y -1

```python
plt.figure(figsize=(10, 5))
plt.plot(trayectoria, color='blue', linewidth=1)
plt.axhline(0, color='red', linestyle='--', alpha=0.5) # Línea de origen
plt.title(f"Caminata Aleatoria 1D (p={p}, q={q:.2f})")
plt.xlabel("Número de pasos")
plt.ylabel("Posición")
plt.grid(True, alpha=0.3)
plt.show()
```

esta parte del codigo es la encargada de tomar los datos simulados y graficar
