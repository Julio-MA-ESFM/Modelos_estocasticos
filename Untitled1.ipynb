```python
import numpy as np
import matplotlib.pyplot as plt

p = 0.5          # Probabilidad de ir a la derecha
n_pasos = 1000   # Número de pasos
q = 1 - p        # Probabilidad de ir a la izquierda

# Generardor de pasos aleatorios
pasos = np.random.choice([1, -1], size=n_pasos, p=[p, q])

# Calculo de pasos acumulados
trayectoria = np.cumsum(pasos)
trayectoria = np.insert(trayectoria, 0, 0) # Añadir el origen

#  Gráfica de  resultados
plt.figure(figsize=(10, 5))
plt.plot(trayectoria, color='blue', linewidth=1)
plt.axhline(0, color='red', linestyle='--', alpha=0.5) # Línea de origen
plt.title(f"Caminata Aleatoria 1D (p={p}, q={q:.2f})")
plt.xlabel("Número de pasos")
plt.ylabel("Posición")
plt.grid(True, alpha=0.3)
plt.show()
```
