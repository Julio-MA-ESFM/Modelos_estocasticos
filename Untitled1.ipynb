```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Parámetros configurables
p = 0.6          # Probabilidad de ir a la derecha (puedes cambiar este valor)
n_pasos = 1000   # Número de pasos
q = 1 - p        # Probabilidad de ir a la izquierda

# 2. Generar los pasos aleatorios
# Generamos n_pasos valores: 1 con prob p, y -1 con prob q
pasos = np.random.choice([1, -1], size=n_pasos, p=[p, q])

# 3. Calcular la trayectoria acumulando los pasos
# Iniciamos en la posición 0
trayectoria = np.cumsum(pasos)
trayectoria = np.insert(trayectoria, 0, 0) # Añadir el origen

# 4. Graficar el resultado
plt.figure(figsize=(10, 5))
plt.plot(trayectoria, color='blue', linewidth=1)
plt.axhline(0, color='red', linestyle='--', alpha=0.5) # Línea de origen
plt.title(f"Caminata Aleatoria 1D (p={p}, q={q:.2f})")
plt.xlabel("Número de pasos")
plt.ylabel("Posición")
plt.grid(True, alpha=0.3)
plt.show()
```
